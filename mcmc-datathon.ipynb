```python
import requests
import time
import datetime

# Your OpenWeatherMap API key (replace with your actual key)
API_KEY = '4b99694e7478e60c63af7e4e3ee4a55c'

# Function to make API call for historical weather data
def get_historical_weather(lat, lon, start_date, end_date):
    # Initialize an empty list to store the historical data
    weather_data = []

    # Loop over each day in the date range
    for single_date in date_range(start_date, end_date):
        # Convert the date to a Unix timestamp
        timestamp = int(single_date.timestamp())
        
        # API endpoint for historical weather data
        url = f'https://api.openweathermap.org/data/2.5/onecall/timemachine'
        
        # Define parameters for the request
        params = {
            'lat': lat,
            'lon': lon,
            'dt': timestamp,  # Unix timestamp for the date
            'appid': API_KEY,
            'units': 'metric'  # For temperature in Celsius
        }

        try:
            # Make the API request
            response = requests.get(url, params=params)
            response.raise_for_status()  # Check for request errors

            # Parse and store the JSON response
            data = response.json()
            weather_data.append(data)

            # Print or log data for each day
            print(f"Date: {single_date.strftime('%Y-%m-%d')} - Temp: {data['current']['temp']}°C")

        except requests.exceptions.HTTPError as err:
            print(f"HTTP error occurred: {err}")
        except Exception as err:
            print(f"Other error occurred: {err}")
        
        # To avoid hitting rate limits, add a short delay
        time.sleep(1)  # 1-second delay between requests
    
    return weather_data

# Helper function to generate a date range
def date_range(start_date, end_date):
    for n in range(int((end_date - start_date).days) + 1):
        yield start_date + datetime.timedelta(n)

# Example usage
if __name__ == "__main__":
    # Define the location (latitude and longitude for Kuala Lumpur, Malaysia)
    latitude = 1.527549
    longitude = 103.745476

    # Define the date range (from 1 Jan 2023 to today's date)
    start_date = datetime.datetime(2023, 1, 1)
    end_date = datetime.datetime.now()  # Current date

    # Fetch historical weather data for each day in the range
    weather_history = get_historical_weather(latitude, longitude, start_date, end_date)
```
